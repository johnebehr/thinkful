<h1>Research Proposal</h1>

<p style=font-size:.90em;><em>“The Sith Lords are concerned that their recruiting slogan, "Give In to Your Anger," isn't very effective. Darth Vader develops an alternative slogan, "Together We Can Rule the Galaxy." They compare the slogans on two groups of 50 captured droids each. In one group, Emperor Palpatine delivers the "Anger" slogan. In the other, Darth Vader presents the "Together" slogan. 20 droids convert to the Dark Side after hearing Palpatine's slogan, while only 5 droids convert after hearing Vader's. The Sith's data scientist concludes that "Anger" is a more effective slogan and should continue to be used.”</em></p>

<h2>Problem:</h2>
<p>The Galactic Empire needs effective slogans to help drive recruitment in the armed forces.  Presently the slogan “Give in to Your Anger” isn’t proving as effective as was originally believed.  A new slogan has been proposed (“Together We can Rule the Galaxy”).  We are looking for means to determine which slogan is more effective.</p>

<h2>Hypothesis:</h2>
<p>New slogans are an effective means to drive recruitment into the Galactic Empire’s armed forces.  A new recruitment slogan (“Together We can Rule the Galaxy”) should create new interest in the Galactic military. </p>

<h2>Method of testing:</h2>
<p>Draw three groups of 50 unrecruited Galactic citizens each.  Each group should be as broadly representative of recruitment aged citizens as possible.  Group one is the Control group will be presented with recruitment material without either slogan.  Group two will be presented with recruitment material with the current, “Give in to Your Anger” slogan.  Group three will be presented with the new, “Together We can rule the Galaxy” slogan.  All recruitment material will be except for the lack of, or addition of a recruitment slogan.  After the presentation  the citizens will be tracked to determine if they join the Imperial military.  </p>
