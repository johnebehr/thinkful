<h1>Assignments</h1>

<h3>Five components of an A/B test:<h3>
<p>
    <ul>
        <li><strong>Two versions</strong> of something whose effects will be compared</li>
        <li>A <strong>Sample</strong> divided into two groups.</li>
        <li>A <strong>Hypothesis</strong> of what you expect to happen.</li>
        <li><strong>Outcome(s)</strong> of interest.  What is the expected change?</li>
        <li>Other measured variables</li>
    </ul>
</p>

<p>1. <em>Does a new supplement help people sleep better?</em></p>
<p>
<ul>
    <li><strong>Versions</strong>:<br />
        <p>This experiment will seek to compare the quality of sleep experienced by control group versus a treatment group.</p>  
    </li>
    <li><strong>Sample</strong>:<br />
        <p>Care should be taken so that the sample groups are representative of the population at large with consideration given to factors like: age, sex, and ethnicity and large enough to reduce bias.  The experiments's participants will be asked to log and rate their sleep qualiy over specific timeframe (one month, six months, longer).  Only the treatment group will be given the supplement.  The control group will be infomred that they are participating in a qualitative sleep study.</p>
    </li>
    <li><strong>Hypothesis</strong>:<br />
        <p>The expectation is that experiment participants give the supplement will have a better qualitative sleep experience over those who do not take the supplement.</p>
    </li>
    <li><strong>Outcome(s)</strong>:<br />
        <p>The expectation is that study participants that take the supplement will see an improvement in the quality of their sleep.  The key metric for the experiment will be an increase in the score of the sleep quality metric.  The expectation is that that score will increase for those who take the supplement over the course of the study.</p>
    </li>
    <li><strong>Other measured variables</strong>:<br />
        <p>
            <ul>
                <li>Sleep Quality Basline: Taken at the beginning of the experiment and used to compare quality scores against over the course of the study</li>
                <li>Night Over Night Comparison: How did the participants sleep quality compare to the previous night?</li>
                <li>Week Over Week Comparison: How does the participants sleep quality compare to one week ago?</li>
                <li>Sleep Quality at Study End: How does the participants sleep quality comapre to when the study began?</li>
            </ul>
        </p>
    </li>
</ul>
</p>

<p>2. <em>Will new uniforms help a gym's business?</em></p>
<ul>
    <li><strong>Versions</strong>:<br />
        <p>This experiment seeks to determine if new staff uniforms help a gym's business.  The experiment can only be structured as before uniform purchase and after uniform purchase.  The experiment will gauge consumer interaction with the staff before and after the purchase of new uniforms.</p>
    </li>
    <li><strong>Sample</strong>:<br />
        <p>In this experiment the sample group and the population are the same.  The gym is unable to control for who transits through the facility at any given time.</p>    
    </li>
    <li><strong>Hypothesis</strong>:<br />
        <p>With the purchase and deployment of new uniforms the gym anticipates an increase in interaction between staff and consumers.</p>
    </li>
    <li><strong>Outcome(s)</strong>:<br />
        <p>Increased interaction between staff and consumers should lead to an improved customer experience thus improving the gym's business.  Beyond improved customer satisfaction, the gym will be looking to see increased membership signups and rise in the number of interactions between staff and consumers.</p>
    </li>
    <li><strong>Other Measured Variables</strong>:<br />
        <p>In addition to measuring the number of customer interactions, the gym will be montering social media postings for conversations related to the new uniforms.  Additionally, two customer satisfaction surveys will be undertaken before and after the deployment of the new uniforms to check for changes in satisfaction levels before the new uniforms versus after the new uniforms.</p>
    </li>
</ul>

<p>3. <em>Will a new homepage improve my online exotic pet rental business?</em></p>
<ul>
    <li><strong>Versions</strong>:<br />
        <p>This experiment will contrast the current pet rental business homepage to a new version of the homepage.</p>
    </li>
    <li><strong>Sample</strong>:<br />
        <p>Two sample groups can be assembled the control group will be asked to navigate the site using the current homepage.  The treatment group will be asked to navigate the site using the new homepage.  Group participants should be as representative as possible of the general population.  Care should be taken to represent diversity of race, ethnicity, income, and gender and sex.</p>
    </li>
    <li><strong>Hypothesis</strong>:<br />
        <p>Improvements to site usability should translate to increased pet rental business.</p>
    </li>
    <li><strong>Outcomes</strong>:<br />
        <p>The pet rental business is looking to see improved sales from a new homepage.  The new site, with a focus on improved usability and user experience should translate to higher sales revenue.</p>
    </li>
    <li><strong>Other Measured Variables</strong>:<br />
        <p>The pet business will be looking to see improvements in the abiltiy for the user to navigate the site efficently.  Additionally, they will be looking to streamline the order placement process.  They will also be looking to increases to site traffic.</p>
    </li>
</ul>

<p>4. <em>If I put 'please read' in the email subject will more people read my emails?</em></p>
<ul>
    <li><strong>Versions</strong>:<br />
        <p>The control version will contain a Subject Line sans 'please read'.  The treatment version will contain a subject line with 'please read'</p>
    </li>
    <li><strong>Sample</strong>:<br />
        <p>Each version of the email solicitation can bet sent to randomly selected addresses in the subscriber list.  A record will be kept as to who received which version of the email broadcast.</p>
    </li>
    <li><strong>Hypothesis</strong>:<br />
        <p>The addition of 'please read' will encourage more people to, at least, open the email.</p>
    </li>
    <li><strong>Outcomes</strong>:<br />
        <p>The 'please read' request will improve the open rate for the email broadcasts.</p>
    </li>
    <li><strong>Other Measured Variables</strong>:<br />
        <p>In addition to improvements to the open rate the requestor wants to know how much the recipent interacts with links included in the email.</p>
    </li>
</ul>
