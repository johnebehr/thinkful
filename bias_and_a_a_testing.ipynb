<h1>Bias and A/A testing</h1>
<h2>Assignments</h2>

<p style=font-size:1.25em;>1. <em>You're testing advertising emails for a bathing suit company and you test one version of the email in February and the other in May.</em>
<p>There are two types of potential bias occuring in this experiment: <em>Sampling bias</em> and <em>Assignment to conditions bias</em>.  Sampling bias is introduced because the sample groups are not representative of the typical customer.  Due to the groups being drawn from from a peek sales month and a low sale month Assignment to conditions bias comes into play.  This assumes that the designers of the test are not looking specifically for these customers for another purpose.  The test could be staged to use past emails, updated for the use case, in an attempt to control for the seasonal effects.</p>

<p style=font-size:1.25em;>2. <em>You open a clinic to treat anxiety and find that the people who visit show a higher rate of anxiety than the general population.</em></p>
<p>The clinic is failing to consider <em>Assignment to conditions bias</em> since they specialized in anxiety treatment visitors will self-selct to visit the clinic thus reducing the randomness of clinic visitors.  The self-selcting nature of the clients skews the sample population (<em>Sampling bias</em>) further causing the study to be non-normative.</p>

<p style=font-size:1.25em;>3. <em>You launch a new ad billboard based campaign and see an increase in website visits in the first week.</em></p>
<p>This experiment has fallen victim to an <em>Assignment to conditions bias</em>.  If the increased web traffic was an unexpected outcome for the experiment then the study designers failed to consider the impact of non-internet base advertising.  In the alternative, they failed to control for the outside factor.</p>

<p style=font-size:1.25em;>4. <em>You launch a loyalty program but see no change in visits in the first week.</em></p>
<p>This study appears to suffer from <em>Observer bias</em> and <em>Assignment to conditions bias</em>.  The experiment assumes a much faster uptick in customer visits than is realisticly possible.  The experiment is assuming too short a timeframe causing an <em>Assignment to conditions bias</em> and an <em>Observer bias</em> where the observer has the expectation of an imediate return.   Additionally, the loyalty program depends on the participants joining which introduces a <em>Sampling bias</em>.  </p>
