# Data distributions normal and otherwise

```python
import statistics
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt

# Set a global for the display tags
tag_params = dict(boxstyle='round',ec='gainsboro',fc='whitesmoke')
mean_color = 'green'
stdev_color = 'red'

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 11,
        }

mean_fnt = dict(color=mean_color)
stdev_fnt = dict(color=stdev_color)
```

```python
rndm_binomial = np.random.binomial(n=10,p=.5,size=100)
binomial_mean = rndm_binomial.mean()
binomial_stdev = rndm_binomial.std()

plt.hist(rndm_binomial)
plt.axvline(x=binomial_mean,color=mean_color)
_, max_ = plt.ylim()

binomial_mean_tag = dict(x=binomial_mean+1, y=max_-max_/10,s=f'Mean: {binomial_mean:.2f}')
plt.text(x=binomial_mean_tag['x'],y=binomial_mean_tag['y'],s=binomial_mean_tag['s'],fontdict=mean_fnt,bbox=tag_params)
plt.axvline(x=binomial_mean+binomial_stdev,color=stdev_color)
_, max_ = plt.ylim()

binomial_stdev_positive_tag = dict(x=(binomial_mean+binomial_stdev)+1, 
                                   y=max_-max_/5 ,
                                   s=f'StDev: {binomial_mean+binomial_stdev:.2f}')

plt.text(x=binomial_stdev_positive_tag['x'], 
         y=binomial_stdev_positive_tag['y'], 
         s=binomial_stdev_positive_tag['s'],
         fontdict=stdev_fnt, bbox=tag_params)
plt.axvline(x=binomial_mean-binomial_stdev,color=stdev_color)
_, max_ = plt.ylim()
binomial_stdev_negative_tag = dict(x=(binomial_mean-binomial_stdev)+1, 
                                   y=max_-max_/5, 
                                   s=f'StDev: {binomial_mean-binomial_stdev:.2f}')

plt.text(x=binomial_stdev_negative_tag['x'], 
         y=binomial_stdev_negative_tag['y'], 
         s=binomial_stdev_negative_tag['s'], 
         fontdict=stdev_fnt, bbox=tag_params)

plt.title('Binomial Distribution of a randomly generated variable', fontdict=font)
# plt.xlabel('X axis', fontdict=font)
# plt.ylabel('Y axis', fontdict=font)

plt.show()
```

```python
# pareto
rndm_pareto = np.random.pareto(3,100)
pareto_mean = np.mean(rndm_pareto)
pareto_stdev = statistics.stdev(rndm_pareto)

plt.hist(rndm_pareto)

plt.axvline(x=pareto_mean, color=mean_color)
_, max_ = plt.ylim()
pareto_mean_tag = dict(x=pareto_mean+.16, y=max_-max_/10, s=f'Mean: {pareto_mean:.2f}')
plt.text(x=pareto_mean_tag['x'], y=pareto_mean_tag['y'], s=pareto_mean_tag['s'], fontdict=mean_fnt, bbox=tag_params)

plt.axvline(x=pareto_mean+pareto_stdev, color=stdev_color)
_, max_ = plt.ylim()
pareto_stdev_positive_tag = dict(x=(pareto_mean+pareto_stdev)+.16, y=max_-max_/5, s=f'StDev: {pareto_mean+pareto_stdev:.2f}')
plt.text(x=pareto_stdev_positive_tag['x'], 
         y=pareto_stdev_positive_tag['y'], 
         s=pareto_stdev_positive_tag['s'], 
         fontdict=stdev_fnt, bbox=tag_params)

plt.axvline(x=pareto_mean-pareto_stdev, color=stdev_color)
_, max_ = plt.ylim()
pareto_stdev_positive_tag = dict(x=(pareto_mean-pareto_stdev)+.16, y=max_-max_/5, s=f'StDev: {pareto_mean-pareto_stdev:.2f}')
plt.text(x=pareto_stdev_positive_tag['x'], 
         y=pareto_stdev_positive_tag['y'], 
         s=pareto_stdev_positive_tag['s'], 
         fontdict=stdev_fnt, bbox=tag_params)

plt.title('Pareto Distribution of a randomly generated variable', fontdict=font)
# plt.xlabel('X axis', fontdict=font)
# plt.ylabel('Y axis', fontdict=font)

plt.show()
```

```python
# poisson
rndm_poisson = np.random.poisson(5, 100)
poisson_mean = np.mean(rndm_poisson)
poisson_stdev = statistics.stdev(rndm_poisson)

plt.hist(rndm_poisson)

plt.axvline(x=poisson_mean, color=mean_color)
_, max_ = plt.ylim()
poisson_mean_tag = dict(x=poisson_mean+.20, y=max_-max_/10, s=f'Mean: {poisson_mean:.2f}')
plt.text(x=poisson_mean_tag['x'], y=poisson_mean_tag['y'], s=poisson_mean_tag['s'], fontdict=mean_fnt, bbox=tag_params)

plt.axvline(x=poisson_mean+poisson_stdev, color=stdev_color)
_, max_ = plt.ylim()
poisson_possitive_stdev_tag = dict(x=(poisson_mean+poisson_mean)-2.25, y=max_-max_/5, s=f'StDev: {poisson_mean+poisson_stdev:.2f}')
plt.text(x=poisson_possitive_stdev_tag['x'], 
         y=poisson_possitive_stdev_tag['y'], 
         s=poisson_possitive_stdev_tag['s'], 
         fontdict=stdev_fnt, 
         bbox=tag_params)

plt.axvline(x=poisson_mean-poisson_stdev, color=stdev_color)
_, max_ = plt.ylim()
poisson_negative_stdev_tag = dict(x=(poisson_mean-poisson_mean)+3.25, y=max_-max_/5, s=f'StDev: {poisson_mean-poisson_stdev:.2f}')
plt.text(x=poisson_negative_stdev_tag['x'], 
         y=poisson_negative_stdev_tag['y'], 
         s=poisson_negative_stdev_tag['s'], 
         fontdict=stdev_fnt, 
         bbox=tag_params)

plt.title('Poisson Distribution of a randomly generated variable', fontdict=font)
# plt.xlabel('X axis', fontdict=font)
# plt.ylabel('Y axis', fontdict=font)

plt.show()
```
