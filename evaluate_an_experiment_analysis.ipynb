<h1>Challenge: Evaluate an experiment analysis</h1>

<p style=font-size:0.9em;>1. <em>The Sith Lords are concerned that their recruiting slogan, "Give In to Your Anger," isn't very effective. Darth Vader develops an alternative slogan, "Together We Can Rule the Galaxy." They compare the slogans on two groups of 50 captured droids each. In one group, Emperor Palpatine delivers the "Anger" slogan. In the other, Darth Vader presents the "Together" slogan. 20 droids convert to the Dark Side after hearing Palpatine's slogan, while only 5 droids convert after hearing Vader's. The Sith's data scientist concludes that "Anger" is a more effective slogan and should continue to be used.</em></p>
<p>I assume that the data scientists in charge of the experiment are aware of who Emperor Palpatine is and therefore I would argue that signifigant case can be made for <em>Observer bias</em>.  Additionally, droids are not suitable test subjects since they are incapable of moral constructs which is a pretty clear case of <em>Selection bias</em>.</p>

<p style=font-size:0.9em;>2. <em>In the past, the Jedi have had difficulty with public relations. They send two envoys, Jar Jar Binks and Mace Windu, to four friendly and four unfriendly planets respectively, with the goal of promoting favorable feelings toward the Jedi. Upon their return, the envoys learn that Jar Jar was much more effective than Windu: Over 75% of the people surveyed said their attitudes had become more favorable after speaking with Jar Jar, while only 65% said their attitudes had become more favorable after speaking with Windu. This makes Windu angry, because he is sure that he had a better success rate than Jar Jar on every planet. The Jedi choose Jar Jar to be their representative in the future.</em></p>
<p>First the experiment suffers from <em>Sampling bias</em> because the two sample groups lack randomness.  The experiment is also flaweed when it comes to <em>Assignment to conditions bias</em> because all of sample one is friendly to the Jedi cause and all of sample is unfriendly to the Jedi cause.  <em>Context bias</em> is also involved from the standpoint that each planet is going to have a miriad of differences from the other planets which would be very difficult to control for.</p>


<p style=font-size:0.9em;>3. <em>A company with work sites in five different countries has sent you data on employee satisfaction rates for workers in Human Resources and workers in Information Technology. Most HR workers are concentrated in three of the countries, while IT workers are equally distributed across worksites. The company requests a report on satisfaction for each job type. You calculate average job satisfaction for HR and for IT and present the report.</em></p>
<p>In order to try produce a viable report I would like to draw samples for HR and IT from countries where both groups are represented.  Further, I'd like to review the demographics of the employees to make sure that they are sufficiently diverse.  A review of the survey questions would need to be undertaken to make sure that the questions are as culture and ethnically, neutral as possible.</p>

<p style=font-size:0.9em;>4. <em>When people install the Happy Days Fitness Tracker app, they are asked to "opt in" to a data collection scheme where their level of physical activity data is automatically sent to the company for product research purposes. During your interview with the company, they tell you that the app is very effective because after installing the app, the data show that people's activity levels rise steadily.</em></p>
<p>The glaring problem here is that the users have self-selected to participate.  By extension a reasonable assumption can be made that it is in the users self-interest to increase their activity level.</p>

<p style=font-size:0.9em;>5. <em>To prevent cheating, a teacher writes three versions of a test. She stacks the three versions together, first all copies of Version A, then all copies of Version B, then all copies of Version C. As students arrive for the exam, each student takes a test. When grading the test, the teacher finds that students who took Version B scored higher than students who took either Version A or Version C. She concludes from this that Version B is easier, and discards it.</em></p>
<p>There is a possible <em>Sampling bias</em> problem here, the teacher does not have sufficient assurance that the tests are being randomly distributed.  Additionaly, are the sample groups of the same size.  Is the teacher drawing her conclusion from the entire population or is she drawing samples from the tests?  This could be an example of Simpson's paradox.</p>
