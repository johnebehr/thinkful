```python
def t_vals(inputs):
    numerator = inputs['y1'] - inputs['y2']
    denominator = math.sqrt((math.pow(inputs['s1'],2)/inputs['n1'])+(math.pow(inputs['s2'],2)/inputs['n2']))
    t = numerator/denominator
    return t
```

```python
# 1. 
inputs_1 = dict(y1=5, y2=8, s1=1, s2=3, n1=200, n2=500)
t1 = t_vals(inputs_1)

print(t1)
```
